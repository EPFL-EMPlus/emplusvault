```python
scores['harmonized_sport'] = scores['sport'].map(sport_mapping)
```

```python
# Convert 'na' in 'end_date' to the corresponding value in 'start_date'
scores.loc[scores['end_date'] == 'na', 'end_date'] = scores.loc[scores['end_date'] == 'na', 'start_date']

# Convert date columns to datetime format, ignoring 'na' values
metadata['date'] = pd.to_datetime(metadata['date'], errors='coerce')
scores['start_date'] = pd.to_datetime(scores['start_date'], errors='coerce')
scores['end_date'] = pd.to_datetime(scores['end_date'], errors='coerce')

# Initialize an empty DataFrame to store the merged data
merged_df = pd.DataFrame()

# Iterate through each row in the metadata DataFrame
for idx, row in metadata.iterrows():
    sport = row['sport']
    date = row['date']

    # Skip rows where date is NaT (Not a Time)
    if pd.isna(date):
        continue

    # Filter the scores DataFrame based on the sport and date range, ignoring NaT
    filtered_scores = scores[(scores['harmonized_sport'] == sport) & 
                             (scores['start_date'].notna()) & 
                             (scores['end_date'].notna()) & 
                             (scores['start_date'] <= date) & 
                             (scores['end_date'] >= date)]

    # If matches are found, merge the rows
    if not filtered_scores.empty:
        # Create a DataFrame that duplicates the current row in metadata for each match found
        duplicated_metadata = pd.DataFrame([row]*filtered_scores.shape[0], columns=metadata.columns)

        # Merge and append to the merged_df
        merged_data = pd.concat([duplicated_metadata.reset_index(drop=True), filtered_scores.reset_index(drop=True)], axis=1)
        merged_df = pd.concat([merged_df, merged_data])

# Reset the index for the final merged DataFrame
merged_df.reset_index(drop=True, inplace=True)
```

```python
# Step 1: Preprocess the scores DataFrame
scores_start = scores.copy()
scores_start['event_date'] = pd.to_datetime(scores['start_date'], errors='coerce')

scores_end = scores.copy()
scores_end['event_date'] = pd.to_datetime(scores['end_date'], errors='coerce')

# Filter out rows where 'event_date' is NaT
scores_start = scores_start[scores_start['event_date'].notna()]
scores_end = scores_end[scores_end['event_date'].notna()]

# Concatenate the start and end date DataFrames
scores_concat = pd.concat([scores_start, scores_end])

# Filter out rows where 'event_date' is NaT, again to be safe
scores_concat = scores_concat[scores_concat['event_date'].notna()]

# Step 1: Filter only final rounds in metadata (adapt this based on how 'final' is indicated in your DataFrame)
metadata_finals = metadata[metadata['round'].str.contains('Final', case=False, na=False)]

# Step 2: Sort both DataFrames by date
metadata_finals = metadata_finals.sort_values(by='date')
scores_concat = scores_concat.sort_values(by='event_date')

# Step 3: Use merge_asof to perform the optimized merge
merged_df = pd.merge_asof(metadata_finals, scores_concat, by='sport', left_on='date', right_on='event_date', direction='backward')

# Step 4: Filter where date falls within the start and end date range
merged_df = merged_df[(merged_df['date'] >= merged_df['start_date']) & (merged_df['date'] <= merged_df['end_date'])]

# Step 5: Drop redundant columns (specify the columns you want to keep)
# columns_to_keep = ['guid', 'seq_id', 'date', 'duration', 'sport', 'round', 'result_id', 'event_title', 'edition', 'result_location', 'result_participants', 'result_description']
# merged_df = merged_df[columns_to_keep]
```
